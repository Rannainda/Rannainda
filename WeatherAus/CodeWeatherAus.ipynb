```python
df = df.drop(["Evaporation","Sunshine",'Cloud9am', 'Cloud3pm'], axis =1)
```

```python
# Xóa những dòng có giá trị null trong cột "RainTomorrow"
df=df.dropna(subset=['RainTomorrow'])

# In thông tin về dữ liệu sau khi xóa
print(df.info())
```

```python
# Replace numerical columns with median
def replace_numerical(df):
    for col in df.select_dtypes(['int', 'float']):
        df[col] = df[col].fillna(df[col].median())
    return df

# Replace object columns with mode
def replace_object(df):
    for col in df.select_dtypes('object'):
        df[col] = df[col].fillna(method='ffill')
    return df
```

```python
df = replace_numerical(df)
df = replace_object(df)
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_cols = ['Date',	'Location',	'WindGustDir',	'WindDir9am',	'WindDir3pm',	'RainToday',	'RainTomorrow']

# mã hóa dữ liệu
ordinal_encoder = OrdinalEncoder()
df[ordinal_cols] = ordinal_encoder.fit_transform(df[ordinal_cols])


# thêm dữ liệu đã mã hóa vào df ban đầu và xóa những cột chưa mã hóa
df_encoded = pd.concat([df[ordinal_cols], df.drop(columns=ordinal_cols)], axis=1)
print(df_encoded)
```

```python
from sklearn.model_selection import train_test_split
X= df_encoded.drop(columns=['RainTomorrow'])
Y= df_encoded["RainTomorrow"]
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size= 0.2, random_state=29)
print(X_train.shape, X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
model_scale = StandardScaler()
model_scale.fit(X_train)
X_train_scale = model_scale.transform(X_train)
X_test_scale = model_scale.transform(X_test)
print(X_train)
print(X_train_scale)
```
